## Фильтрация

```python
bikes['Holiday'] == 'Holiday'   # праздничные дни - 
```

```python
bikes[bikes['Holiday'] == 'Holiday']   # праздничные дни - список ссотв. строк
```

```python
# праздничные дни и температура > 15C
bikes[(bikes['Holiday'] == 'Holiday') & (bikes['Temperature'] > 15)]
```

```python
# праздничные дни: лето ИЛИ и температура > 25C
bikes[(bikes['Seasons'] == 'Summer') | (bikes['Temperature'] > 25)]
```

```python
# праздничные дни, КРОМЕ: лето ИЛИ и температура > 25C
bikes[~((bikes['Seasons'] == 'Summer') | (bikes['Temperature'] > 25))]
```

```python
# пропущено значение температуры
bikes[bikes['Temperature'].isna()]
```

```python
# Время года -  в списке ['Winter', 'Autamn']
bikes[bikes['Seasons'].isin(['Winter', 'Autamn'])]
```

```python
# Температура -  в интервале (-10,5)
bikes[bikes['Temperature'].isin(range(-10, 6))]
```
