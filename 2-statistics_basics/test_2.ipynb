### Какой регион проживания реже всего встречается в данных?

```python
import collections

frequency_by_region = collections.Counter(df['loc'])
frequency_by_region
```

```python
given_locs = ['Северо-Восточный центральный', 'Тихоокеанский', 'Южно-Атлантический',
              'Средне-Атлантический', 'Горный', 'Новая Англия']

for loc in given_locs:
    print(f"{loc} - {frequency_by_region[loc]}")
```
