### Верно ли следующее утверждение: мужчины 18-29 лет, которые не являются фанатами серии, чаще всего проживают в Тихоокеанском регионе.

```python
filtered_df = df[(df['gender'] == 'М') & (df['age_group'] == '18-29') & (df['fan'] == 'нет')]
filtered_df
```

```python
filtered_df['loc'].value_counts()

# Ответ: верно
```
